Now that we don't have to deal with 30K Na rows, the dataset looks a little more manageable. The next step here is to clean the 'Date' column, so we create a new column called 'Proper_date' which is easier to manage.

```python
shark_data_unclean['Time'] = shark_data_unclean['Time'].replace(regex='([a-zA-Z])', value=0)
shark_data_unclean['Time'] = shark_data_unclean['Time'].fillna(0)
#shark_data_unclean['Time'] = shark_data_unclean['Time'].astype(int)

#Activity
shark_data_unclean['Activity'] = shark_data_unclean['Activity'].fillna(0)

#Proper Date
shark_data_unclean['Proper_date'] = shark_data_unclean['Proper_date'].fillna(0)
shark_data_unclean['Proper_date'] = shark_data_unclean['Proper_date'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['Proper_date'] > 19000000] 
average = int(shark_data_date["Proper_date"].mean())
shark_data_unclean["Proper_date"].replace({0: average}, inplace=True)
print(average)

#Age
shark_data_unclean['Age'] = shark_data_unclean['Age'].fillna(0)
shark_data_unclean['Age'] = shark_data_unclean['Age'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['Age'] > 0] 
average = int(shark_data_date["Age"].mean())
shark_data_unclean["Age"].replace({0: average}, inplace=True)
print(average)


#avg_temp
shark_data_unclean['avg_temp'] = shark_data_unclean['avg_temp'].fillna(0)
#shark_data_unclean['avg_temp'] = shark_data_unclean['avg_temp'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['avg_temp'] > 0] 
average = shark_data_date["avg_temp"].mean()
shark_data_unclean["avg_temp"].replace({0: average}, inplace=True)
print(average)


#precipitation
shark_data_unclean['precipitation'] = shark_data_unclean['precipitation'].fillna(0)
#shark_data_unclean['precipitation'] = shark_data_unclean['precipitation'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['precipitation'] > 0] 
average = shark_data_date["precipitation"].mean()
shark_data_unclean["precipitation"].replace({0: average}, inplace=True)
print(average)


#wind_speed
shark_data_unclean['wind_speed'] = shark_data_unclean['wind_speed'].fillna(0)
#shark_data_unclean['wind_speed'] = shark_data_unclean['wind_speed'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['wind_speed'] > 0] 
average = shark_data_date["wind_speed"].mean()
shark_data_unclean["wind_speed"].replace({0: average}, inplace=True)
print(average)


#sea_level_pressure
shark_data_unclean['sea_level_pressure'] = shark_data_unclean['sea_level_pressure'].fillna(0)
#shark_data_unclean['sea_level_pressure'] = shark_data_unclean['sea_level_pressure'].astype('int')


shark_data_date = shark_data_unclean[shark_data_unclean['sea_level_pressure'] > 0] 
average = shark_data_date["sea_level_pressure"].mean()
shark_data_unclean["sea_level_pressure"].replace({0: average}, inplace=True)
print(average)
shark_data_unclean
```
