```python
import datetime
from datetime import date
import urllib
import ssl
import csv
import codecs
import json

ssl._create_default_https_context = ssl._create_unverified_context


def getDate(input_date):
    print(input_date)
    try:
        d = datetime.datetime.strptime(input_date, "%d-%b-%Y")
        
    except:
        try:
            d = datetime.datetime.strptime(input_date, "%d-%b%Y")
        except:
            try:
                d = datetime.datetime.strptime(input_date, " %d-%b-%Y")
            except:
                try:
                    d = datetime.datetime.strptime(input_date, "%d-%b-%Y ")
                except:
                    try:
                        d = datetime.datetime.strptime(input_date, "%b-%Y")
                        return "del"
                    except:
                        try:
                            d = datetime.datetime.strptime(input_date, "%b %Y")
                            return "del"
                            
                        finally:
                            print()
                    finally:
                        print()
                finally:
                     print()
            finally:
                print()
        finally:
            print()
    finally:   
        output_date = d.strftime("%Y-%m-%d")
        return output_date
        
#Code taken from visualcrossing.com
def get_weather_data(lat, lng, day):
    # This is the core of our weather query URL
    BaseURL = 'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/weatherdata/'

    # Set up the location parameter for our query
    QueryLocation = '&location=' + lat + ',' + lng

    # Set up the query type parameter for our query ('FORECAST' or 'HISTORY')
    QueryType='HISTORY'

    # Set up the key parameter for our query
    QueryKey = '&key=' + '6NFPPC92EKQ4JX2W2XX8JMAFN'

    # Set up the date parameters for our query. Used only for historical weather data requests
    FromDateParam = day
    ToDateParam = day
    
    # Set up the specific parameters based on the type of query
    print(' - Fetching history for date: ', FromDateParam,'-',ToDateParam)

    # History requests require a date.  We use the same date for start and end since we only want to query a single date in this example
    QueryDate = '&startDateTime=' + FromDateParam + 'T00:00:00&endDateTime=' +ToDateParam + 'T00:00:00'
    QueryTypeParams = 'history?&aggregateHours=24&unitGroup=us&dayStartTime=0:0:00&dayEndTime=0:0:00' + QueryDate


    # Build the entire query
    URL = BaseURL + QueryTypeParams + QueryLocation + '&extendedStats=True&' + QueryKey + "&contentType=json" + "&maxDistance=200000"

    print(' - Running query URL: ', URL)
    print()

    # Parse the results as CSV
    response = urllib.request.urlopen(URL)
    json_data = response.read()
    data = json.loads(json_data.decode('utf-8'))
    
    loc_data = list(data["locations"])[0]
    dict_data = data["locations"][loc_data]["values"] 
    return (dict_data[0]["temp"], dict_data[0]["precip"], dict_data[0]["mean_wspd"], dict_data[0]["sealevelpressure"])

for index, item in shark_data.iterrows():
    if item["na_wind_speed"] == True:
        cur_date = getDate(item["Proper_date"])
        if(cur_date != "del"):
            data_tuple = get_weather_data(str(item["latitude"]), str(item["longitude"]), cur_date)
            print(data_tuple)
            shark_data["avg_temp"][index] = data_tuple[0]
            shark_data["precipitation"][index] = data_tuple[1]
            shark_data["wind_speed"][index] = data_tuple[2]
            shark_data["sea_level_pressure"][index] = data_tuple[3]
            shark_data["na_wind_speed"][index] = False
```
